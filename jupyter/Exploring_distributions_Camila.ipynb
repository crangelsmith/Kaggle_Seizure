```python
import glob
import os
import scipy.io
import pandas
from highcharts import Highchart
from scipy.signal import resample
import matplotlib.pyplot as plt
import numpy as np
```

```python
pwd = str(os.getcwd())
# data location, relative.
dir_ = '/../data/train_1/'
# define test file, layout is from when i wanted to iterate over all files
# to run over all files change line to '*.mat'
file_type_0 = '1_78_0.mat'
file_type_1 = '1_78_1.mat'

file_0 = glob.glob(pwd+dir_+file_type_0)
file_1 = glob.glob(pwd+dir_+file_type_1)

print file_0
print file_1
```

```python
mat_0 = scipy.io.loadmat(str(file_0[0]))
mat_1 = scipy.io.loadmat(str(file_1[0]))

headers = ['channel0', 'channel1', 'channel2', 'channel3',
                   'channel4', 'channel5', 'channel6', 'channel7',
                   'channel8', 'channel9', 'channel10', 'channel11',
                   'channel12', 'channel13', 'channel14', 'channel15']

        # get actual data from the .mat file
channels_data_0 = mat_0['dataStruct'][0][0][0]
channels_data_1 = mat_1['dataStruct'][0][0][0]
```

```python
rs_data_0 = resample(channels_data_0, 3600, axis=0)
rs_data_1 = resample(channels_data_1, 3600, axis=0)

df_0 = pandas.DataFrame(rs_data_0, columns=headers)
df_1 = pandas.DataFrame(rs_data_1, columns=headers)
```
